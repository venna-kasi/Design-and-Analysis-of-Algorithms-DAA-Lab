```python
import time

def bubble_sort(arr):
    n = len(arr)
    for i in range(n - 1):
        # Last i elements are already in place
        for j in range(0, n - i - 1):
            # Traverse the array from 0 to n-i-1
            # Swap if the element found is greater than the next element
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

# Get user input for the list
user_input = input("Enter numbers separated by spaces for Bubble Sort: ")
try:
    my_list_bubble = list(map(int, user_input.split()))
except ValueError:
    print("Invalid input. Please enter numbers only.")
    my_list_bubble = []

if my_list_bubble:
    print(f"Original list: {my_list_bubble}")

    # Measure execution time
    start_time = time.time()
    sorted_list_bubble = bubble_sort(list(my_list_bubble)) # Use a copy to avoid modifying original list before sort
    end_time = time.time()
    execution_time = end_time - start_time

    print(f"Sorted list (Bubble Sort): {sorted_list_bubble}")
    print(f"Execution Time (Bubble Sort): {execution_time:.6f} seconds")
    print("Time Complexity (Bubble Sort): O(N^2) in average/worst case, O(N) in best case")
    print("Space Complexity (Bubble Sort): O(1) auxiliary space")
else:
    print("No valid numbers entered for Bubble Sort.")
```

```python
import time

def selection_sort(arr):
    n = len(arr)
    for i in range(n - 1):
        # Assume the current index has the minimum element
        min_index = i

        # Find the index of the minimum element
        for j in range(i + 1, n):
            if arr[j] < arr[min_index]:
                min_index = j

        # Swap the found minimum element with the first unsorted element
        arr[i], arr[min_index] = arr[min_index], arr[i]

    return arr

# Get user input for the list
user_input = input("Enter numbers separated by spaces for Selection Sort: ")

try:
    my_list_selection = list(map(int, user_input.split()))
except ValueError:
    print("Invalid input. Please enter numbers only.")
    my_list_selection = []

if my_list_selection:
    print(f"Original list: {my_list_selection}")

    # Measure execution time
    start_time = time.time()
    sorted_list_selection = selection_sort(list(my_list_selection))  # Use a copy
    end_time = time.time()
    execution_time = end_time - start_time

    print(f"Sorted list (Selection Sort): {sorted_list_selection}")
    print(f"Execution Time (Selection Sort): {execution_time:.6f} seconds")
    print("Time Complexity (Selection Sort): O(N^2) in best, average, and worst case")
    print("Space Complexity (Selection Sort): O(1) auxiliary space")
else:
    print("No valid numbers entered for Selection Sort.")
```

```python
import time

def merge_sort(arr):
    if len(arr) > 1:
        # Find the middle of the array
        mid = len(arr) // 2

        # Divide the array into two halves
        left_half = arr[:mid]
        right_half = arr[mid:]

        # Recursively sort both halves
        merge_sort(left_half)
        merge_sort(right_half)

        # Merge the sorted halves
        i = j = k = 0

        while i < len(left_half) and j < len(right_half):
            if left_half[i] < right_half[j]:
                arr[k] = left_half[i]
                i += 1
            else:
                arr[k] = right_half[j]
                j += 1
            k += 1

        # Copy remaining elements of left_half
        while i < len(left_half):
            arr[k] = left_half[i]
            i += 1
            k += 1

        # Copy remaining elements of right_half
        while j < len(right_half):
            arr[k] = right_half[j]
            j += 1
            k += 1

    return arr

# Get user input for the list
user_input = input("Enter numbers separated by spaces for Merge Sort: ")

try:
    my_list_merge = list(map(int, user_input.split()))
except ValueError:
    print("Invalid input. Please enter numbers only.")
    my_list_merge = []

if my_list_merge:
    print(f"Original list: {my_list_merge}")

    # Measure execution time
    start_time = time.time()
    sorted_list_merge = merge_sort(list(my_list_merge))  # Use a copy
    end_time = time.time()
    execution_time = end_time - start_time

    print(f"Sorted list (Merge Sort): {sorted_list_merge}")
    print(f"Execution Time (Merge Sort): {execution_time:.6f} seconds")
    print("Time Complexity (Merge Sort): O(N log N) in best, average, and worst case")
    print("Space Complexity (Merge Sort): O(N)")
else:
    print("No valid numbers entered for Merge Sort.")
```

```python
import time

def quick_sort(arr):
    if len(arr) <= 1:
        return arr

    # Choose the last element as the pivot
    pivot = arr[-1]

    # Partition the array
    left = [x for x in arr[:-1] if x <= pivot]
    right = [x for x in arr[:-1] if x > pivot]

    # Recursively sort and combine
    return quick_sort(left) + [pivot] + quick_sort(right)

# Get user input for the list
user_input = input("Enter numbers separated by spaces for Quick Sort: ")

try:
    my_list_quick = list(map(int, user_input.split()))
except ValueError:
    print("Invalid input. Please enter numbers only.")
    my_list_quick = []

if my_list_quick:
    print(f"Original list: {my_list_quick}")

    # Measure execution time
    start_time = time.time()
    sorted_list_quick = quick_sort(list(my_list_quick))  # Use a copy
    end_time = time.time()
    execution_time = end_time - start_time

    print(f"Sorted list (Quick Sort): {sorted_list_quick}")
    print(f"Execution Time (Quick Sort): {execution_time:.6f} seconds")
    print("Time Complexity (Quick Sort): O(N log N) average/best case, O(N^2) worst case")
    print("Space Complexity (Quick Sort): O(log N) average (recursive stack), O(N) worst case")
else:
    print("No valid numbers entered for Quick Sort.")
```
